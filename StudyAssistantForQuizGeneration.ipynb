```python
import PyPDF2
```

```python
# Replace 'your_file.pdf' with the name of your uploaded PDF file
file_path = 'PromptEngineering.pdf'

# Initialize an empty string to hold the extracted text
extracted_text = ""

try:
    # Open the PDF file in binary read mode
    with open(file_path, 'rb') as file:
        # Create a PdfReader object
        reader = PyPDF2.PdfReader(file)

        # Get the total number of pages
        num_pages = len(reader.pages)
        print(f"The PDF has {num_pages} pages.")

        # Loop through each page and extract text
        for page_num in range(num_pages):
            page = reader.pages[page_num]
            extracted_text += page.extract_text() + "\n"  # Add a newline for readability

    # Print the extracted text
    print("\n--- Extracted Text ---")
    print(extracted_text)

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from langchain.prompts import PromptTemplate
```

```python
SYS_PROMPT = """
Act as an instructor in Prompt Engineering.
Summarize the study material loaded in key bullet points.
Do not exceed more than 10 bullet points.
After that generate 4 multiple choice questions, based on the bullet points.
These questions should test a student knowledge on these bullet points.
Each question should have 4 options, and only 1 answer should be correct.

Study Material : {extracted_text}
"""

prompt = PromptTemplate(
    template=SYS_PROMPT,
    input_variables=["extracted_text"]
)
```

```python
chain = (prompt | gemini)
```

```python
response = chain.invoke({"extracted_text": extracted_text})
```
